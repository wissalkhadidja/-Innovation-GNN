```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# Générer une série chronologique de données
num_samples = 100
time_steps = np.arange(num_samples)
data = np.sin(time_steps * 0.1) + np.random.normal(0, 0.1, size=num_samples)

# Créer un modèle CNN simple
model = tf.keras.Sequential([
    layers.Input(shape=(num_samples, 1)),
    layers.Conv1D(filters=16, kernel_size=3, activation='relu', padding='same'),
    layers.Conv1D(filters=1, kernel_size=3, activation=None, padding='same')
])

# Compiler le modèle
model.compile(optimizer='adam', loss='mse')

# Entraîner le modèle (ceci est un exemple simplifié, pas pour la détection d'anomalies)
model.fit(data.reshape(1, num_samples, 1), data.reshape(1, num_samples, 1), epochs=50)

# Utiliser le modèle pour la détection d'anomalies
predicted_data = model.predict(data.reshape(1, num_samples, 1)).squeeze()
seuil_anomalies = 0.5  # Définir un seuil pour les anomalies
anomalies = np.abs(predicted_data - data) > seuil_anomalies

# Afficher les résultats
print("Données Originales:", data)
print("Données Prédites:", predicted_data)
print("Anomalies:", anomalies)

plt.figure(figsize=(10, 6))
plt.plot(time_steps, data, label="Série Chronologique")
plt.plot(time_steps, predicted_data, label="Prédictions")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.title("Série Chronologique et Prédictions")
plt.legend()
plt.show()
```
