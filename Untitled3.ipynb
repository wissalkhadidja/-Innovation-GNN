```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt
# Générer une série chronologique de données
num_samples = 100
time_steps = np.arange(num_samples)
data = np.sin(time_steps * 0.1) + np.random.normal(0, 0.1, size=num_samples)

# Créer un graphe temporel (simple ici, avec des arêtes entre chaque paire de nœuds)
edges = [(i, i+1) for i in range(num_samples-1)]

# Construire la matrice d'adjacence du graphe
adjacency_matrix = np.zeros((num_samples, num_samples))
for edge in edges:
    adjacency_matrix[edge[0], edge[1]] = 1

# Normaliser la matrice d'adjacence pour la propagation de l'information
row_sums = adjacency_matrix.sum(axis=1)
normalized_adjacency_matrix = adjacency_matrix / row_sums[:, np.newaxis]

# Convertir la matrice d'adjacence en tenseur TensorFlow
normalized_adjacency_tensor = tf.constant(normalized_adjacency_matrix, dtype=tf.float32)

# Créer un modèle GNN simple
class GraphConvLayer(layers.Layer):
    def __init__(self, num_output_features):
        super(GraphConvLayer, self).__init__()
        self.num_output_features = num_output_features

    def call(self, inputs):
        features, adjacency = inputs
        aggregated_features = tf.linalg.matmul(adjacency, features)
        return aggregated_features

inputs = layers.Input(shape=(num_samples, 1))
x = GraphConvLayer(num_output_features=16)((inputs, normalized_adjacency_tensor))
x = layers.Activation("relu")(x)
x = GraphConvLayer(num_output_features=1)((x, normalized_adjacency_tensor))
model = tf.keras.Model(inputs=inputs, outputs=x)

# Compiler le modèle et l'entraîner (ceci est un exemple simplifié, pas pour la détection d'anomalies)
model.compile(optimizer='adam', loss='mse')
model.fit(data.reshape(1, num_samples, 1), data.reshape(1, num_samples, 1), epochs=50)

# Utiliser le modèle pour la détection d'anomalies
predicted_data = model.predict(data.reshape(1, num_samples, 1)).squeeze()
seuil_anomalies = 0.5  # Définir un seuil pour les anomalies
anomalies = np.abs(predicted_data - data) > seuil_anomalies

# Afficher les résultats
print("Données Originales:", data)
print("Données Prédites:", predicted_data)
print("Anomalies:", anomalies)

plt.figure(figsize=(10, 6))
plt.plot(time_steps, data, label="Série Chronologique")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.title("Série Chronologique de Données")
plt.legend()
plt.show()
```
